# Clustring on the Housing DataSet
    

```python
import matplotlib.pyplot as  plt
import numpy as np
import pandas as pd 
from sklearn.cluster import KMeans
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv("housing.csv")
print(df.head())
dfheaders = df.columns.tolist()
print(df.dtypes)
data = df.copy()
```

```python
encod = LabelEncoder()
data['ocean_proximity'] = encod.fit_transform(data['ocean_proximity'])
x = np.nan_to_num(data)
```

```python
pd.plotting.scatter_matrix(data[dfheaders],figsize=(17,9))
plt.show()
```

```python
K = np.arange(3,30)
b = np.array([])
for k in K:
    model = KMeans(n_init=10,n_clusters=k).fit(x)
    b = np.append(b,model.inertia_)

    
```

```python
plt.plot(K,b)
```

```python
model = KMeans(n_init=10,n_clusters=10).fit(x)
label = model.labels_
plt.scatter(x[:,0],x[:,-2],c=label)
```

```python
fig = plt.figure(figsize=(16,9))
ax = plt.axes(projection ="3d")
ax.scatter3D(x[:,-2],x[:,3],x[:,0],c=label)
```
